```python
# Celda 1: Importaciones
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Celda 2: Descarga de datos para análisis
ticker = "AAPL" # Usaremos Apple como ejemplo para el EDA
data = yf.download(ticker, period="5y")


# Celda 3: Inspección básica (Gana 1 punto aquí)
print(data.info())
print(data.describe())
print("Valores nulos:", data.isnull().sum())

# Celda 4: Análisis de la Variable Objetivo (Gana el 2do punto)
plt.figure(figsize=(12,6))
plt.title(f"Distribución de Precios de Cierre - {ticker}")
sns.histplot(data['Close'], kde=True, color='blue')
plt.show()

# Celda 5: Análisis de Volatilidad (Retornos diarios)
data['Returns'] = data['Close'].pct_change()
plt.figure(figsize=(12,6))
plt.title("Análisis de Retornos Diarios (Volatilidad)")
sns.boxplot(x=data['Returns'].dropna())
plt.show()

# Celda 6: Correlación (Feature Importance)
plt.figure(figsize=(10,8))
sns.heatmap(data.corr(), annot=True, cmap='coolwarm')
plt.title("Matriz de Correlación de Variables")
plt.show()
